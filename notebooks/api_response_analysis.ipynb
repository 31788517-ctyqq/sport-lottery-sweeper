# 处理示例 API 响应 JSON 并解析其中的数据

此笔记本演示如何加载并分析后端返回的角色列表 JSON 响应。

## 1. 导入必要的库

导入 `json`、`pandas` 等用于处理和显示数据的库。

```python
import json
import pandas as pd

# 示例 JSON 响应字符串
response_json = '{"data":[{"name":"pr1_role_1772341922","description":"pr1 smoke","permissions":[],"id":1,"created_at":"2026-03-01T05:12:02","updated_at":"2026-03-01T05:12:02"}],"total":1,"skip":0,"limit":100}'
```

```python
# 2. 加载 JSON 响应

data_obj = json.loads(response_json)
```

```python
# 3. 解析 `data` 字段
data_list = data_obj.get('data', [])
data_list
```

```python
# 4. 提取名称、描述等信息
df = pd.DataFrame([{
    'name': item.get('name'),
    'description': item.get('description'),
    'id': item.get('id'),
    'permissions_count': len(item.get('permissions', []))
} for item in data_list])
df
```

```python
# 5. 处理分页信息
total = data_obj.get('total')
skip = data_obj.get('skip')
limit = data_obj.get('limit')
print(f"total={total}, skip={skip}, limit={limit}")
# 计算是否需要下一页
if skip + limit < total:
    print("还有下一页")
else:
    print("无下一页")
```

```python
# 6. 将结果打印或可视化
print(df)

# 柱状图示例
if not df.empty:
    df.set_index('name')['permissions_count'].plot(kind='bar')
```
